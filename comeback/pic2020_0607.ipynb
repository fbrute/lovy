## Choose station

```python
filename = filenames[0]
```

## Choose station

```python
station_df = pd.read_csv(filename, sep=';')
station_df['id'] = station_df.index
print(len(station_df.index))
station_df.head()
```

```python
lambdas = ['440', '1020']
```

## Choose lambda

```python
lambda_val = lambdas[0]
```

## Select a minimum of 2 observations (nobs => N[{lambda_val}nm]  with pic2020_0607)

```python
#station_df = station_df[station_df['nobs'] >= 2]
station_df = station_df[station_df[f"N[AOD_{lambda_val}nm]"] >= 2]
print(len(station_df.index))
station_df.head()
```

## Select columns : year, month, days, id and nobs

```python
ymd_df = station_df.filter(regex=f"yyyy|mm|dd|id|N[AOD_{lambda_val}nm]", axis = 1)
ymd_df.head()
```

```python
aod_nobs_df = station_df.filter(regex=f"id|N[AOD_{lambda_val}nm]", axis = 1)
result_df = ymd_df.merge(aod_nobs_df, on='id', how='inner')
result_df.head()
```

## Add a column to result_df

```python
def add_columns(regexp, source_df, result_df):
    source2_df = source_df.copy()
    result2_df = result_df.copy()
    columns_df = source2_df.filter(regex=regexp, axis = 1)
    result_df = columns_df.merge(result2_df, on='id', how='inner')
```

## AOD lambda_val

```python
lambda_df_AOT_avg_std = station_df.filter(regex=f"^AOD_{lambda_val}|id", axis = 1)
result_df = ymd_df.merge(lambda_df_AOT_avg_std, on='id', how='inner')
result_df.head()
```

```python
lambda_df = station_df.filter(regex=f"^AOD_{lambda_val}_med|id", axis = 1)
result_df = result_df.merge(lambda_df, on='id', how='inner')
result_df.head()
```

## 440-870_Angstrom_Exponent

```python
lambda_df = station_df.filter(regex=f"^440-870_Angstrom_Exponent|id", axis = 1)
print(lambda_df.head())
#result_df = result_df.merge(lambda_df, on='id', how='inner')
#result_df.head()
```

```python
lambda_df = station_df.filter(regex=f"870-440AngstromParam,\[AOTExt\]-Total_med|id", axis = 1)
#print(lambda_df.head())
result_df = result_df.merge(lambda_df, on='id', how='inner')
result_df.head()
```

## Inflection_Point

```python
lambda_df = station_df.filter(regex=f"Inflection_Point\[um\]_[avg$|std$]|id", axis = 1)
#print(lambda_df.head())
result_df = result_df.merge(lambda_df, on='id', how='inner')
result_df.head()
```

```python
lambda_df = station_df.filter(regex=f"Inflection_Point\[um\]_med|id", axis = 1)
#print(lambda_df.head())
result_df = result_df.merge(lambda_df, on='id', how='inner')
result_df.head()
```
